```python
import pandas as pd

df = pd.read_csv("./data/mushrooms.csv")

df.head()
```

```python
from sklearn.metrics import precision_score
from sklearn.model_selection import train_test_split, cross_validate
from sklearn import preprocessing
import numpy as np

#Instanziierung des LabelEncoders

le = preprocessing.LabelEncoder()

#Für jede Spalte wird der Labelencoder mithilfe einer for-Schleife angewandt.

#Encode..
for col in df.columns:
    df[col] = le.fit_transform(df[col])
    
#Entfernen der Klasse "class" aus X-Matrix

X = df.drop(columns="class")


#Edible = 0, Poisonous = 1

y = df["class"]


#Aufteilung der Trainings-und Testdaten

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42, test_size = 0.25, stratify = y)


X_train.shape

```

```python
#Erstellen eines Baseline-Modells. Ein Baseline Modell liefert einen ersten Grundwert in Bezug auf Genauigkeit

#Edible = 0, Poisonous = 1

df["class"].value_counts(normalize=True)


```

```python
majority_class = y_train.mode()[0]

baseline_predictions = [majority_class] * len(y_train)
```

```python
#Measuring baseline model accurary 

from sklearn.metrics import accuracy_score

majority_class_accuracy = accuracy_score(baseline_predictions, y_train)

print(majority_class_accuracy)

#Accuracy of baseline model is...0.517971442639094
```

```python
#Create now a Model which defeats the baseline model in prediction

#Decision Tree Classifier

from sklearn.tree import DecisionTreeClassifier
import graphviz
from sklearn.tree import export_graphviz

#Create a new Object called "tree" 

tree = DecisionTreeClassifier(max_depth=1)

#Fit the model

tree.fit(X_train, y_train)

```
